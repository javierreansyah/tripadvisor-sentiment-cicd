```python
import pandas as pd

df = pd.read_csv("../Data/data.csv")
df
```

```python
from datetime import datetime
import pytz

# Get current time in WIB (Western Indonesian Time, UTC+7)
wib = pytz.timezone('Asia/Jakarta')
current_wib_time = datetime.now(wib).strftime('%Y-%m-%d %H:%M:%S')

# Set all values in 'Timestamp' to current WIB time
df['Timestamp'] = current_wib_time

# Move 'Timestamp' to the start
cols = ['Timestamp'] + [col for col in df.columns if col != 'Timestamp']
df = df[cols]
df.head()
```

```python
df.to_csv("../Data/original_data.csv", index=False)
```

```python
df_sampled = df.sample(n=500, random_state=42)
```

```python
df_sampled.to_csv("../Data/new_data.csv", index=False)
```
